```python
import pandas as pd
from tabulate import tabulate
import os
from datetime import datetime

print(os.getcwd())
```

```python
os.chdir('raw_data')
os.chdir('overview')
path = os.getcwd()
print(path)
```

```python
# change file names to date ranges only
files = [ f for f in os.listdir(path) if os.path.isfile(os.path.join(path,f)) ]
for f in files:
    os.rename(os.path.join(path,f),os.path.join(path, f.replace('Analytics Master Overview ', '').lower())) 
    #print (f)
```

```python
files = [ f for f in os.listdir(path) if os.path.isfile(os.path.join(path,f)) ]
df = pd.DataFrame()
week = 1
for f in files:
    temp = pd.read_csv(f, skiprows=6, skipfooter=18, engine='python')
    temp = temp.drop(columns=['Users', 'New Users', 'Bounce Rate', 'Pages / Session', 'Link Click - Apply (Goal 13 Conversion Rate)', 'Link Click - Apply (Goal 13 Value)']) # remove unnecessary columns
    temp.columns = ['Device', 'Sessions', 'Avg_Duration', 'Grad_Apply'] # shorten column names
    
    # add date range from file name
    temp['ID'] = week
    week = week + 1
    temp['Begin Date'] = datetime.strptime(f[0:8], '%Y%m%d').date()
    temp['End Date'] = datetime.strptime(f[9:17], '%Y%m%d').date()
    df = pd.concat([df, temp])
    
```

```python
print(tabulate(df, showindex=False, headers=df.columns))
```

```python
# Split df into separate DataFrames based on 'Device'
device_groups = df.groupby('Device')

# Create separate DataFrames for each device
device_dfs = []
for device, group in device_groups:
    device_dfs.append(group.copy())

# Access individual DataFrames for each device
desktop_df = device_dfs[0]  # DataFrame for 'desktop' device
#desktop_df = desktop_df.drop(columns=['Device'])
mobile_df = device_dfs[1]   # DataFrame for 'mobile' device
#mobile_df = mobile_df.drop(columns=['Device'])
tablet_df = device_dfs[2]   # DataFrame for 'tablet' device
#tablet_df = tablet_df.drop(columns=['Device'])
```

```python
print(tabulate(desktop_df, showindex=False, headers=desktop_df.columns))
```

```python
print(tabulate(mobile_df, showindex=False, headers=mobile_df.columns))
```
